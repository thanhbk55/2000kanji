```
import requests
from bs4 import BeautifulSoup
import time  # Thêm thư viện time để sử dụng sleep

def split_text(text, max_length=2000):
    result = []
    current_segment = ''

    # Tách văn bản dựa trên các chuỗi \n liên tiếp bằng regex, giữ nguyên ký tự phân cách
    import re
    parts = re.split(r'(\n+)', text)  # Tách và giữ bất kỳ số lượng \n nào

    for i, part in enumerate(parts):
        # Nếu phần tử là chuỗi \n (có thể là \n, \n\n, \n\n\n, v.v.)
        if re.match(r'^\n+$', part):
            if len(current_segment) > 0:
                current_segment += part  # Thêm ký tự phân cách vào đoạn hiện tại
            continue

        # Nếu thêm phần mới vượt quá max_length
        if (len(current_segment + part) > max_length) and len(current_segment) > 0:
            result.append(current_segment.rstrip())  # Thêm đoạn hiện tại vào kết quả
            current_segment = part  # Bắt đầu đoạn mới
        else:
            # Nếu chưa vượt quá, thêm phần vào đoạn hiện tại
            if len(current_segment) == 0:
                current_segment = part
            else:
                # Thêm ký tự phân cách trước đó (nếu có)
                if i > 0 and re.match(r'^\n+$', parts[i - 1]):
                    current_segment += parts[i - 1] + part
                else:
                    current_segment += part

    # Thêm đoạn cuối cùng nếu còn
    if len(current_segment) > 0:
        result.append(current_segment.rstrip())

    return result

def fetch_and_parse(url):
    try:
        # Fetch HTML
        response = requests.get(url)
        response.raise_for_status()  # Kiểm tra lỗi HTTP
        data = response.text

        # Load vào BeautifulSoup
        soup = BeautifulSoup(data, 'html.parser')
        content = soup.select_one('#content').get_text()  # Lấy nội dung trong #content

        # Chia đoạn văn bản
        segments = split_text(content, 3000)  # Giới hạn 3000 như trong code của bạn

        # Danh sách để lưu các đoạn đã được viết lại
        rewritten_segments = []

        # Chuẩn bị dữ liệu gửi API và xử lý từng đoạn
        for segment in segments:
            data1 = {
                "contents": [{
                    "parts": [{
                        "text": f"Viết lại theo tiếng Việt Nam chuẩn ngữ pháp, phong cách truyện tiên hiệp: {segment}"
                    }]
                }]
            }

            # Gửi yêu cầu tới API
            api_url = "https://generativelanguage.googleapis.com/v1beta/models/gemini-2.0-flash:generateContent?key=AIzaSyD1937V0dzY2NXXlCmECpHQIGXv4gxoKJs"
            response = requests.post(api_url, json=data1, headers={'Content-Type': 'application/json'})
            response.raise_for_status()  # Kiểm tra lỗi HTTP

            # Lấy nội dung viết lại từ phản hồi API
            rewritten_text = response.json()['candidates'][0]['content']['parts'][0]['text']
            rewritten_segments.append(rewritten_text)

            # Thêm sleep 2 giây giữa các yêu cầu
            time.sleep(2)  # Có thể điều chỉnh thời gian (giây) tùy theo nhu cầu

        # Nối các đoạn đã viết lại thành một văn bản hoàn chỉnh
        final_text = ''.join(rewritten_segments)
        print("Đoạn văn hoàn chỉnh:")
        print(final_text)

    except Exception as error:
        print("Lỗi:", error)

# Gọi hàm
fetch_and_parse("https://truyenconvert.net/doc-truyen/chuong-1-tam-tram-nam-sau-1-4354")
```
